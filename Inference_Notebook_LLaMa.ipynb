### Setup

```python
from peft import PeftModel
from transformers import LlamaTokenizer, LlamaForCausalLM, GenerationConfig

tokenizer = LlamaTokenizer.from_pretrained("baffo32/decapoda-research-llama-7B-hf")

model = LlamaForCausalLM.from_pretrained(
    "baffo32/decapoda-research-llama-7B-hf",
    load_in_8bit=True,
    device_map="auto",
)
model = PeftModel.from_pretrained(model, "kunchum/capstone-llama-finetuned")
```

```python
# Function to generate model response for a given prompt
def generate_model_response(prompt, context=None):
    if context:
        PROMPT = f'''Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.
        ### Instruction:
        {prompt}
        ### Input:
        {context}
        ### Response:
        '''
    else:
        PROMPT = f'''Below is an instruction that describes a task. Write a response that appropriately completes the request.
        ### Instruction:
        {prompt}
        ### Response:
        '''

    inputs = tokenizer(
        PROMPT,
        return_tensors="pt",
    )
    input_ids = inputs["input_ids"].cuda()

    generation_config = GenerationConfig(
        do_sample=True,
        temperature=0.6,
        top_p=0.95,
        repetition_penalty=1.15,
    )

    print("Generating Response...\n")
    generation_output = model.generate(
        input_ids=input_ids,
        generation_config=generation_config,
        return_dict_in_generate=True,
        output_scores=True,
        max_new_tokens=256,
    )

    response = ""
    for s in generation_output.sequences:
        response = tokenizer.decode(generation_output.sequences[0], skip_special_tokens=True)

    # Truncate the response at the first occurrence of triple quotes (""")
    if '"""' in response:
        response = response.split('"""')[0].strip()
        response = response.split('### Response:')[1].strip()

    return response
```

## Rouge Score Evaluation

```python
from datasets import load_dataset, DatasetDict
RANDOM_SEED = 1234

# Load the dataset
data = load_dataset('kunchum/capstone_1')
data = data.shuffle(seed=RANDOM_SEED)  # Shuffle dataset

# Sample 20,000 records for training
train_sample_size = 20000
data_sample = data['train'].select(range(train_sample_size))

# Ensure no overlap between train and test
remaining_data = data['train'].select(range(train_sample_size, len(data['train'])))

# Sample 1,000 records for testing from the remaining data
test_sample_size = 1000
test_sample = remaining_data.select(range(test_sample_size))

# Create a DatasetDict to save train and test splits
final_data_dict = DatasetDict({
    'train': data_sample,
    'test': test_sample
})

# Save the dataset to disk
final_data_dict.save_to_disk("train_test_dataset")
```

```python
import pandas as pd
from evaluate import load
from tqdm import tqdm
import re
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

# Load ROUGE scorer
rouge_scorer = load("rouge")

# Load test data
test_samples = final_data_dict["test"]
num_samples_to_evaluate = 5
samples = test_samples.select(range(num_samples_to_evaluate))

# Evaluate model responses and compute ROUGE scores
results = []

print("Generating responses and computing ROUGE scores...")

def custom_evaluation(ground_truth, model_response):
    clarity_score = evaluate_clarity(ground_truth, model_response)
    specificity_score = evaluate_specificity(ground_truth, model_response)
    relevance_score = evaluate_relevance(ground_truth, model_response)

    return clarity_score, specificity_score, relevance_score

# Clarity Evaluation: Measures if the advice is clear and easy to understand
def evaluate_clarity(ground_truth, model_response):
    # Simple heuristic: compare the sentence length (simpler = clearer)
    ground_truth_length = len(ground_truth.split())
    model_response_length = len(model_response.split())
    clarity_score = max(0, 1 - abs(ground_truth_length - model_response_length) / max(1, ground_truth_length))
    return clarity_score

# Specificity Evaluation: Measures how specific and detailed the advice is
def evaluate_specificity(ground_truth, model_response):
    # Simple heuristic: calculate the number of financial terms used in the response
    financial_terms = ['interest rate', 'credit score', 'investment', 'loan', 'debt', 'tax', 'dividends', 'savings', 'risk', 'budget', 'inflation', 'mortgage', 'retirement fund', 'stocks', 'bonds', 'mutual funds', 'real estate', '401k', 'IRA', 'financial planning', 'cash flow', 'portfolio', 'asset allocation', 'liquidity', 'capital gains', 'expense ratio', 'net worth', 'emergency fund', 'spending limit', 'income tax', 'estate planning', 'savings account', 'checking account', 'payroll', 'tax deduction', 'tax bracket', 'diversification',  'credit card debt', 'financial independence', 'living expenses', 'financial advisor',  'liabilities', 'assets', 'wealth management', 'pension', 'recession', 'financial literacy']
    ground_truth_terms = sum(1 for term in financial_terms if term in ground_truth.lower())
    model_response_terms = sum(1 for term in financial_terms if term in model_response.lower())
    specificity_score = min(1, model_response_terms / (ground_truth_terms + 1))  # Avoid division by 0
    return specificity_score

# Relevance Evaluation: Measures if the response answers the original financial question
def evaluate_relevance(ground_truth, model_response):
    # Using cosine similarity to measure relevance (this can be more complex using BERT embeddings or other NLP techniques)
    vectorizer = TfidfVectorizer().fit([ground_truth, model_response])
    tfidf_matrix = vectorizer.transform([ground_truth, model_response])
    similarity_matrix = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
    relevance_score = similarity_matrix[0][0]
    return relevance_score

for sample in tqdm(samples):
    test_prompt = sample["instruction"]
    test_context = sample["context"]
    ground_truth = sample["response_cleaned"]

    # Generate model response
    if test_context:
        model_response = generate_model_response(test_prompt, test_context)
    else:
        model_response = generate_model_response(test_prompt)

    # Compute ROUGE scores
    scores = rouge_scorer.compute(predictions=[model_response], references=[ground_truth])

    clarity, specificity, relevance = custom_evaluation(ground_truth, model_response)

    # Store results in a dictionary
    results.append({
        "Test Set Response": ground_truth,
        "Model Response": model_response,
        "ROUGE-L Score": scores["rougeL"],
        "ROUGE-1 Score": scores["rouge1"],
        "ROUGE-2 Score": scores["rouge2"],
        "Clarity Score": clarity,
        "Specificity Score": specificity,
        "Relevance Score": relevance
    })

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
```
